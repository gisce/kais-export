```python
df = pd.read_excel('traspaso_Informacion_de campos.xls')
df.drop([0,1,2,3], inplace=True)
df.head()
```

```python
fields_df = []
for idx, row in df.iterrows():
    field, tipus = row[0].split()
    field = field.lower()
    tipus = tipus.replace(',;', '').lower()
    if tipus == "d":
        field_declaration = 'fields.DateField(length=8, format="%Y%m%d")'
    elif '(' in tipus:
        tipus, length = tipus.split('(')
        length = length.replace(')', '')
        if tipus == 'c':
            field_declaration = 'fields.StringField(length={})'.format(length)
        elif tipus == 'n':
            if ',' in length:
                length, decimals = length.split(',')
                field_declaration = 'fields.DecimalField(length={}, decimals={})'.format(length, decimals)
            else:
                field_declaration = 'fields.IntegerField(length={})'.format(length)
    fields_df.append({
        'field': field,
        'field_declaration': field_declaration,
        'length': int(length)
    })
fields_df = pd.DataFrame(fields_df)
fields_df.head()
```

```python
code = []
for idx, row in fields_df.iterrows():
    code.append("    {field} = {field_declaration}".format(**row))

print('\n'.join(code))
```
